```python
import pandas as pd

# Veri seti yükleme
data_path = r"C:\Users\Mehmet\Desktop\archive\creditcard.csv"
df = pd.read_csv(data_path)

# İlk 5 satır
df.head()
```

```python
# Genel bilgi
df.info()

# Eksik veri var mı?
df.isnull().sum()

# Sınıf dağılımı (Fraud / Not Fraud)
df['Class'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df['Amount_scaled'] = scaler.fit_transform(df[['Amount']])
df['Time_scaled'] = scaler.fit_transform(df[['Time']])

# Artık orijinal 'Amount' ve 'Time' sütunlarına ihtiyacımız yok:
df = df.drop(['Amount', 'Time'], axis=1)

df.head()
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print("Eğitim seti boyutu:", X_train.shape)
print("Test seti boyutu:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import mlflow
import mlflow.sklearn

mlflow.set_experiment("CreditCardFraudDetection")

with mlflow.start_run():

    # Model tanımlama ve eğitim
    model = LogisticRegression(max_iter=1000)
    model.fit(X_train, y_train)

    # Tahmin
    y_pred = model.predict(X_test)

    # Metrikler
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # MLflow Loglama
    mlflow.log_param("model_type", "LogisticRegression")
    mlflow.log_param("max_iter", 1000)
    
    mlflow.log_metric("accuracy", acc)
    mlflow.log_metric("precision", prec)
    mlflow.log_metric("recall", rec)
    mlflow.log_metric("f1_score", f1)

    # Modeli logla
    mlflow.sklearn.log_model(model, "model")

print(f"Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import mlflow
import mlflow.sklearn

mlflow.set_experiment("CreditCardFraudDetection")

with mlflow.start_run():

    # Model tanımlama
    clf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
    clf.fit(X_train, y_train)

    # Tahmin
    y_pred = clf.predict(X_test)

    # Metrikler
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # MLflow Loglama
    mlflow.log_param("model_type", "RandomForestClassifier")
    mlflow.log_param("n_estimators", 100)
    mlflow.log_param("max_depth", 10)

    mlflow.log_metric("accuracy", acc)
    mlflow.log_metric("precision", prec)
    mlflow.log_metric("recall", rec)
    mlflow.log_metric("f1_score", f1)

    # Model kaydı
    mlflow.sklearn.log_model(clf, "model")

print(f"RandomForestClassifier -> Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [5, 10],
    'min_samples_split': [2, 5]
}

rf = RandomForestClassifier(random_state=42)

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           scoring='f1', cv=3, verbose=2, n_jobs=-1)

grid_search.fit(X_train, y_train)

print("En iyi parametreler:", grid_search.best_params_)
print("En iyi F1 Skoru:", grid_search.best_score_)
```
